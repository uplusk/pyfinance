1: Manually download monthly and daily price data for CitiGroup

```python
citi_daily = pd.read_csv('C.csv')
citi_monthly = pd.read_csv('C (1).csv')
```

```python
citi_daily.head()
citi_daily['Date'] = citi_daily['Date'].apply(pd.to_datetime)
citi_monthly['Date'] = citi_monthly['Date'].apply(pd.to_datetime)
```

```python
citi_daily.set_index('Date',inplace=True)
citi_monthly.set_index('Date',inplace=True)
```

2: Convert daily price data for the CitiGroup to daily returns

```python
citi_daily['Daily returns'] = citi_daily['Adj Close'].pct_change(1)
citi_monthly['Monthly returns'] = citi_monthly['Adj Close'].pct_change(1)
```

```python
citi_daily['Daily returns'].head()
```

3: Convert monthly prices to monthly returns and convert daily returns to
monthly returns. Are they the same

```python
citi_monthly['Monthly returns'].head()
```

```python
# Converting daily data to monthly 
```

```python
monthly = citi_daily.resample(rule='MS').sum()['Daily returns']
```

```python
monthly.head()
```

```python
citi_monthly['Monthly returns'].plot(title='Comparing monthly returns',label='Monthly',figsize=(12,8))
monthly.plot(label='Daily-Monthly')
plt.legend()
```

```python
#Yes the monthly returns and the daily converted monthly returns are exactly the same.
```

4: Download IBM monthly data from Yahoo Finance. Estimate its standard
deviation and Sharpe ratio from January 2000 to December 2004.

```python
ibm = pd.read_csv('IBM.csv')
```

```python
ibm.dropna(inplace=True)
```

```python
ibm.head()
```

```python
ibm.set_index(pd.to_datetime(ibm['Date']),inplace=True)
```

```python
ibm[ibm.index<'2005-01-01']['Adj Close'].std()
```

```python
ibm[ibm.index<'2005-01-01']['Adj Close'].plot(figsize=(12,8))
```

```python
monthly_ret = ibm[ibm.index<'2005-01-01']['Adj Close'].pct_change(1)
```

```python
monthly_ret.hist(bins=100,figsize=(12,8))
```

```python
sharpe_ratio = monthly_ret.mean()/monthly_ret.std()
```

```python
sharpe_ratio*(252**0.5)
```

```python
vedanta_stock['MA50'] = vedanta_stock['Open'].rolling(50).mean()
vedanta_stock['MA200'] = vedanta_stock['Open'].rolling(200).mean()

nmdc_stock['MA50'] = nmdc_stock['Open'].rolling(50).mean()
nmdc_stock['MA200'] = nmdc_stock['Open'].rolling(200).mean()

maithan_stock['MA50'] = maithan_stock['Open'].rolling(50).mean()
maithan_stock['MA200'] = maithan_stock['Open'].rolling(200).mean()
```

```python
vedanta_stock['Returns'] = vedanta_stock['Close'].pct_change(1)
nmdc_stock['Returns'] = nmdc_stock['Close'].pct_change(1)
maithan_stock['Returns'] = maithan_stock['Close'].pct_change(1)
```

```python
box_df = pd.concat([vedanta_stock['Returns'], nmdc_stock['Returns'], maithan_stock['Returns']],axis =1)
box_df.columns = ['VEDL Returns','NMDC Returns','MAITHANALL Returns']
box_df.plot(kind='box',figsize=(8,10),colormap='jet')
```

```python
vedanta_stock['Cum Return'] = (1+vedanta_stock['Returns']).cumprod()
nmdc_stock['Cum Return'] = (1+nmdc_stock['Returns']).cumprod()
maithan_stock['Cum Return'] = (1+maithan_stock['Returns']).cumprod()
```

```python
vedanta_stock['Cum Return'].plot(label='VEDL Returns',figsize=(16,8))
nmdc_stock['Cum Return'].plot(label='NMDC Returns')
maithan_stock['Cum Return'].plot(label='MAITHANALL Returns')
plt.legend()
```

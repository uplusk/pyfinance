Monte Carlo Simulation for lets say 15000 random allocations

```python
num_ports = 15000
weights_arr = np.zeros((num_ports,len(all.columns)))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sr_arr = np.zeros(num_ports)

for i in range(num_ports):
    
    weights = np.array(np.random.random(4))
    
    #Rebalancing
    weights = weights/np.sum(weights)
    
    weights_arr[i,:] = weights

    #Expected Return
    ret_arr[i] = np.sum(log_ret.mean()*weights)*252


    # Expected Variance
    vol_arr[i] = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov() * 252, weights)))

    # Sharpe Ratio
    sr_arr[i] = ret_arr[i] - 0.07/vol_arr[i]

```

```python
max_sr_ret = ret_arr[12325]
max_sr_vol = vol_arr[12325]
```

```python
plt.figure(figsize=(16,8))
plt.scatter(vol_arr,ret_arr,c=sr_arr,cmap='plasma')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel = 'Volatility'
plt.ylabel = 'Returns'

#The Bullet Curve
plt.scatter(max_sr_vol,max_sr_ret,c='red',edgecolors='black',s=50)
```

```python
from scipy.optimize import minimize
```

```python
def get_vol_ret_sr(weights):
    weights = np.array(weights)
    ret = np.sum(log_ret.mean() * weights) * 252
    vol = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov() * 252, weights)))
    sr = ret-0.07/vol
    return np.array([ret,vol,sr])
```

```python
def neg_sr(weights):
    return get_vol_ret_sr(weights)[2] * -1                      
```

```python
def check_sum(weights):
    return np.sum(weights) - 1
```

```python
#Constraints
con = ({'type':'eq','fun': check_sum})
```

```python
#Bounds 
bounds = ((0,1),(0,1),(0,1),(0,1))
```

```python
#init_guess
init_guess = [0.25,0.25,0.25,0.25]
```

```python
opt_results = minimize(neg_sr,init_guess,method='SLSQP',bounds=bounds,constraints=con)
```

```python
opt_results
```

```python
opt_results.x
```

```python
get_vol_ret_sr(opt_results.x)
```

### The Optimal Portfolio - Efficient Frontier

```python
frontier_volatility = []
frontier_ret = np.linspace(-0.15,0.15,100)
```

```python
def minimium_volatility(weights):
    return get_vol_ret_sr(weights)[1]
```

```python
for possible_return in frontier_ret:
    #constraints
    cons = ({'type':'eq', 'fun': check_sum},{'type':'eq','fun': lambda w: get_vol_ret_sr(w)[0] - possible_return})
    
    result = minimize(minimium_volatility,init_guess,method='SLSQP',bounds=bounds,constraints=cons)
    frontier_volatility.append(result['fun'])
```

```python
plt.figure(figsize=(16,8))
plt.scatter(vol_arr,ret_arr,c=sr_arr,cmap='plasma')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel = 'Volatility'
plt.ylabel = 'Returns'

#adding frontier line
plt.plot(frontier_volatility,frontier_ret,'g--',linewidth=3)
```
